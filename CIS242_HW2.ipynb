#### 6. Consider the three questions from question #1.  What changes might you need to make to the data to answer them?  Why or why not?

```python
#Three questions that can be asked:
#1. Is there a correlation between mortgage amount and family size?
#2. Are people with more years of professional experience more likely to have credit cards?
#3. Are people less educated more likely to take personal loans?

print(bank.shape)
print(bank.isnull().sum())
#code borrowed from Professor Cannon's CIS 242 Lecture_2.pdf

#As mentioned in the answers to quesiton 1, we know there is no missing data point in this data set.

#1. Is there a correlation between mortgage amount and family size?
#Because this question deals with two numeric varaibles that are not missing any data points, there is no need to make any change.

#2. Are people with more years of professional experience more likely to have credit cards?
#Professional experience is entered as numeric values while credit cards are entered as binary (0 or 1) values.
#Grouping the data points first by years of experience and getting the proportion of individuals who responded 'yes/1' to having
#a credit card will make it easier to see if years of experience has any correlation with having credit cards.
Exp_crd = bank.groupby(['Experience'], as_index=False).mean()
Exp_crd
```

```python
#Years of experience. 
x = Exp_crd[Exp_crd.columns[0]]
#Proportion of people with credit cards.
y = Exp_crd[Exp_crd.columns[12]]
plt.scatter(x, y, color = 'red')
plt.xlabel('Years of experience')
plt.ylabel('Proportion of people with credit cards')
plt.show()

#It appears that there are more people without credit cards across all years of experience with most groups falling in between 0.2 and 0.4, 
#but there doesn't seem to be a correlation between the two variables.
```

```python
#3. Are people more educated more likely to take personal loans?

#Education is entered as ordinal values (1 to 3) while credit cards are entered as binary (0 or 1) values.
#Grouping the data points first by years of experience and getting the proportion of individuals who responded 'yes/1' to taking
#a personal loan will make it easier to see if education level has any correlation with taking personal loans.
Edu_loan = bank.groupby(['Education'], as_index=False).mean()
Edu_loan
```

```python
#Education level.
x = Edu_loan[Edu_loan.columns[0]]
#Propotion of people who took personal loans.
y = Edu_loan[Edu_loan.columns[8]]

plt.bar(x, y, align = 'center', width = 0.5, color = 'green')
plt.xlabel('Education level')
plt.ylabel('Proportion of people with personal loans')
plt.show()

#It appears people with only undergraduate degrees are less likely to take personal loans than people with graduate/advanced degrees.
#This might be attributable to people with advanced degrees having more students loans or having higher rate of loan approval
#due to their education level functioning as higher credibility, although the causal factor remains to be seen.
```
